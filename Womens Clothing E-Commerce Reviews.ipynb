## Data description

This notebooks works with a dataset about women's e-commerce clothing reviews. The Reviews are collected from an anonymous source and is retrieved from: https://www.kaggle.com/datasets/nicapotato/womens-ecommerce-clothing-reviews/data 

<br>

The dataset includes 23486 rows and 10 feature variables: <br>
- *Clothing ID*: Refers to the specific piece being reviewed. 
- Age: The reviewers age.
- Title: The title of the review. 
- Review Text: The review body.
- Rating: The product score granted by the costumor from 1 (worst) to 5 (best).
- Recommended IND: Recommendation from costumer where 1 is recommended and 0 is not recommended. 
- Positive Feedback Count: Documenting the number of other customers who found the specific review positive.
- Division Name: Categories of womens clothing sorted into broader sections (ex. Tendy, Petite, etc.)
- Department Name: Name of the product department (ex. Dresses, department, bottoms, tops, etch )
- Class Name: Name of the product class name.

First of all we take a look at the data types and make sure that the different data has the right data type.
We convert the objects into strings. 

```python
df.info()
```

```python
df["Title"] = df["Title"].astype("string")
df["Review Text"] = df["Review Text"].astype("string")
df["Division Name"] = df["Division Name"].astype("string")
df["Department Name"] = df["Department Name"].astype("string")
df["Class Name"] = df["Class Name"].astype("string")
```

Rename the first column to review ID

```python
df = df.rename(columns={"Unnamed: 0":"Review ID"})
```

```python
df.head()
```

First of all we would like to remove any duplicate rows. The following code will remove any duplicate rows if any. 

```python
df = df.drop_duplicates()
```

We look to see if there are any outliers in the values in the ages of the users.
There seems to be no extreme outliers in the age data

```python
#create 5000 row random sample
sampled_data = df.sample(n=5000, random_state=42)

chart1 = alt.Chart(sampled_data).mark_bar(color="blue").encode(
    x="Age:O",
    y="count():Q"
)

chart1
```

We also want to make sure that the rating data is between 1 and 5.

```python
df["Rating"].min(),df["Rating"].max()
```

Replace the "NA" values with text "no value"

```python
#check if columns has missing values

for column in df.columns:
    has_missing = df[column].isna().any()
    if has_missing:
        print(f"Column '{column}': has missing values.")
    else:
        print(f"Column '{column}': has no missing values.")
```

```python
#replace nan values with appropriate no value text
df['Title'] = df['Title'].fillna('No title')
df['Review Text'] = df['Review Text'].fillna('No review text')
df['Division Name'] = df['Division Name'].fillna('No devision name')
df['Department Name'] = df['Department Name'].fillna('No department name')
df['Class Name'] = df['Class Name'].fillna("No class name")
```

```python
#we want to remove any rows where the title is "No title" and or the Review text is "No review text"
df = df[(df['Title'] != 'No title') & (df['Review Text'] != 'No review text')]
```

**What clothing items have the best and worst average reviews?**

```python
# Filter out rows with 'no class name' in the 'Class Name' column
filtered_data = df[df['Class Name'] != 'No class name']

# Group by 'Class Name' and calculate the mean rating and count
class_ratings = filtered_data.groupby('Class Name')['Rating'].agg(['mean', 'count']).reset_index()

# Rename columns for clarity
class_ratings.columns = ['Class Name', 'Mean Rating', 'Review Count']

# Find the class with the highest and lowest mean ratings
best_item = class_ratings.loc[class_ratings['Mean Rating'].idxmax()]
worst_item = class_ratings.loc[class_ratings['Mean Rating'].idxmin()]

# Sort for better readability
class_ratings = class_ratings.sort_values(by='Mean Rating', ascending=False)

# Display results
print("Number of different 'Class Name' items:", class_ratings.shape[0])
print("\nClass with the Best Mean Rating:")
print(best_item)
print("\nClass with the Worst Mean Rating:")
print(worst_item)
```
